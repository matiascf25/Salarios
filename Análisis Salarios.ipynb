```python
# Tarea, Normalización Z-score
# Importe el archivo a un DataFrame de pandas.
import pandas as pd

df = pd.read_csv("Salary Data.csv")
df.head()
```

```python
# Filtre el DataFrame para mantener solo las columnas de tipo numérico.

col_nums = df.select_dtypes(include="number")
col_nums.head()
```

```python
# Aplique la transformación Z-score.

zscore = (col_nums - col_nums.mean()) / col_nums.std()
zscore.head()
```

```python
# Para el DataFrame estandarizado, calcule y muestre la media (mean()) y desviación estándar (std()) de cada característica.

media = zscore.mean()
desviacion = zscore.std()

print("Media de cada columna:")
print(media)
print("\nDesviación estándar de cada columna:")
print(desviacion)
```
